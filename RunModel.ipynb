```python
def evaluate_results(results: list, labels: list):
    y_sum = 0
    for index, res in enumerate(results):
        y_sum += int(labels[index] == res)

    return f'{(y_sum / len(labels)) * 100}%'
```

```python
is_weighted = True
```

```python
for i in range(1, 30):
  print(f"Iteration with K = {i}")
  knn = KNNWeightedClassifier(i, 'l1', x_train=train_features[0], labels=labels[0], x_data=val_features[0], classes=[0, 1, 2]) if is_weighted else KNNClassifier(i, 'eucledean', x_train=train_features[0], labels=labels[0], x_data=val_features[0], classes=[0, 1, 2])
  results = knn.eval()
  print("Validation", evaluate_results(results, labels=val_labels))
  knn.change_data(test_features[0])
  test_results = knn.eval()
  print("Test", evaluate_results(test_results, labels=test_labels))
  knn.change_data(train_features[0])
  train_results = knn.eval()
  print("Train", evaluate_results(train_results, labels=labels[0]))
  print()


# Seems like k = 8 is the best we can do so far
# K = 3, behaves better is we split the data into train/validation (test is deone from test.csv)
```
